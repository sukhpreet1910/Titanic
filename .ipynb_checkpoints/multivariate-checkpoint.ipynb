```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# Importing Data sets 

train = pd.read_csv('Datasets/train.csv')
flight = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

# 1. Scatterplot (Numeric - Numeric)

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = tips['sex'], style = tips.time, size = tips['size'])
plt.show()
```

# 2. Barplot (Numerical - Categorical)

```python
train.head()
```

```python
sns.barplot(x = 'Pclass', y = 'Fare', data = train, hue = train.Sex)
```

# 3. Boxplot (Numerical - Categorical)

```python
sns.boxplot(x = 'Sex', y = 'Age', data = train, hue = 'Survived')
plt.show()
```

# 4. kdeplot (Numerical - Categorical)

```python
sns.kdeplot(train[train['Survived'] == 0]['Age'])
sns.kdeplot(train[train['Survived'] == 1]['Age'])
plt.show()
```

# 5. Heatmap (Categorical - Categorical)

```python
pd.crosstab(train['Pclass'], train['Survived'])
```

```python
sns.heatmap(pd.crosstab(train['Pclass'], train['Survived']))
```

```python
pd.crosstab(train['Embarked'], train['Survived'])
```

```python
(train.groupby('Embarked')['Survived'].mean()*100)
```

```python
(train.groupby('Embarked')['Survived'].mean()*100).plot(kind = 'bar')
plt.show()
```

# 6. Clustermap (Categorical - Categorical)

```python
sns.clustermap(flight.pivot_table(values='passengers', index='month', columns='year'))
```

```python
(train.groupby('Embarked')['Survived'].mean()*100).plot(kind = 'bar')
plt.show()
```

# 7. Pairplot

```python
sns.pairplot(iris, hue = 'species')
```

# 8. Lineplot (Numerical - Numerical)

```python
new_flight = flight.groupby('year')['passengers'].sum().reset_index()
new_flight
```

```python
sns.lineplot(data = new_flight, x = 'year', y = 'passengers')
plt.show()
```
